```python
import matplotlib as mlp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
siny = y.copy()
cosy = np.cos(x)
```

## 1.plot折线图)

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")  #加以注释
plt.legend()  #自动在图片合适的位置加以图示
plt.axis([-1, 11, -1.5, 1.5])  #修改坐标轴的范围(或者用xlim,ylim)
plt.title("Welcome to the Machine Learning")  #加标题
```

## 2.Scatter Plot(散点图:通常描述二维的特征)

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="green")
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.3)  #alpha表示不透明度
```

```python
plt.scatter(x, y, alpha=0.3, color="red", label="test")
plt.legend()
plt.axis([-5, 5, -5, 5])
plt.xlabel("test X")
plt.ylabel("test Y")
plt.title("test T")
```
